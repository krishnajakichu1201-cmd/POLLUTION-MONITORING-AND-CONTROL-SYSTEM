```python
import pandas as pd

# Load dataset
df = pd.read_csv('/mnt/data/airpollution.csv')

# Remove missing values
df = df.dropna(subset=['AQI'])

# Convert AQI column to numeric (if not already)
df['AQI'] = pd.to_numeric(df['AQI'], errors='coerce')

# Drop rows where AQI conversion failed
df = df.dropna(subset=['AQI'])

# Function to classify AQI
def classify_aqi(aqi):
    if aqi <= 50:
        return 'Good'
    elif aqi <= 100:
        return 'Satisfactory'
    elif aqi <= 200:
        return 'Moderate'
    else:
        return 'Unhealthy'

# Create new column
df['AQI_Category'] = df['AQI'].apply(classify_aqi)

# Keep only required categories
df_cleaned = df[df['AQI_Category'].isin(['Good', 'Satisfactory', 'Moderate'])]

# Reset index
df_cleaned = df_cleaned.reset_index(drop=True)

# Save cleaned dataset
df_cleaned.to_csv('cleaned_air_quality.csv', index=False)

print(df_cleaned.head())
print("done")
```
