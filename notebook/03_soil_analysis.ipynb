```python
land_uses = ['Agricultural','Industrial','Forest','Coastal','Urban']

rows = []
for year in range(2010, 2024):
    for d in districts:
        for lu in land_uses:
            ind = 1.8 if lu == 'Industrial' else 1.0
            ph  = round(np.random.normal(6.2, 0.5), 2)
            ec  = round(np.random.normal(0.45 * ind, 0.1), 3)
            om  = round(np.random.normal(2.1, 0.6), 2)
            lead = round(np.random.normal(18 * ind, 5), 2)
            cd   = round(np.random.normal(0.3 * ind, 0.1), 3)
            cr   = round(np.random.normal(40 * ind, 10), 2)
            pest = round(np.random.normal(0.08, 0.03), 4)
            n    = round(np.random.normal(220, 40), 1)
            p    = round(np.random.normal(18, 5), 1)
            arsenic = round(np.random.normal(8 * ind, 2), 2)
            mercury = round(np.random.normal(0.1 * ind, 0.04), 3)
            moisture = round(np.random.normal(28, 6), 1)
            sqc = 'Good' if (lead<50 and cd<0.5 and ph>5.5) else 'Contaminated'
            rows.append([year, d, lu, 'Kerala', ph, ec, om, lead, cd, cr,
                         pest, n, p, sqc, arsenic, mercury, moisture, n+p])

df_s = pd.DataFrame(rows, columns=['year','district','land_use','state','pH','EC',
    'organic_matter','lead','cadmium','chromium','pesticide','nitrogen','phosphorus',
    'soil_quality_class','arsenic','mercury','moisture','nutrients'])
df_s.to_csv('data/soil_quality.csv', index=False)
```
