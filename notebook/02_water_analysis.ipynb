```python
rivers = ['Periyar','Chaliyar','Bharathapuzha','Pamba',
          'Kallada','Valapattanam','Muvattupuzha']
quarters = ['Q1','Q2','Q3','Q4']

rows = []
for year in range(2010, 2024):
    for river in rivers:
        for q in quarters:
            industrial = 1.4 if river in ['Periyar','Chaliyar'] else 1.0
            bod = round(np.random.normal(4.5 * industrial, 1.2), 2)
            cod = round(bod * 2.8 + np.random.normal(0, 2), 2)
            tds = round(np.random.normal(320 * industrial, 60), 1)
            nitrates = round(np.random.normal(8, 2), 2)
            phosphates = round(np.random.normal(0.4, 0.1), 3)
            fc = int(np.random.normal(1200 * industrial, 300))
            hm = round(np.random.normal(0.05 * industrial, 0.02), 3)
            ph = round(np.random.normal(7.1, 0.4), 2)
            turb = round(np.random.normal(15, 5), 1)
            do_ = round(np.random.normal(6.5, 1.0), 2)
            wq = 'Safe' if (bod < 5 and ph > 6.5 and do_ > 5) else 'Unsafe'
            rows.append([year, q, river, 'Kerala', bod, cod, tds, nitrates,
                         phosphates, fc, hm, wq, ph, turb, do_])

df_w = pd.DataFrame(rows, columns=['year','Quarter','river','state','BOD','COD',
    'TDS','Nitrates','phosphates','fecal_coliform','heavy_metals',
    'water_quality_class','pH','turbidity','dissolved_oxygen'])
df_w.to_csv('data/water_quality.csv', index=False)
```
