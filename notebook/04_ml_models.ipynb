```python
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('data/air_quality.csv')

# 1. AQI trend over years
yearly = df.groupby('year')['AQI'].mean()
plt.figure(figsize=(10,5))
plt.plot(yearly.index, yearly.values, marker='o', color='steelblue')
plt.title('Average AQI trend in Kerala (2010–2023)')
plt.xlabel('Year'); plt.ylabel('AQI')
plt.savefig('outputs/aqi_trend.png', dpi=150, bbox_inches='tight')

# 2. District-wise AQI bar chart
dist_aqi = df.groupby('district')['AQI'].mean().sort_values(ascending=False)
plt.figure(figsize=(12,5))
dist_aqi.plot(kind='bar', color='coral')
plt.title('Average AQI by district')
plt.xticks(rotation=45, ha='right')
plt.savefig('outputs/district_aqi.png', dpi=150, bbox_inches='tight')

# 3. Pollutant correlation heatmap
plt.figure(figsize=(8,6))
sns.heatmap(df[['PM2.5','PM10','NO2','CO','SO2','O3','AQI']].corr(),
            annot=True, cmap='YlOrRd', fmt='.2f')
plt.title('Air pollutant correlation heatmap')
plt.savefig('outputs/air_heatmap.png', dpi=150, bbox_inches='tight')
```
